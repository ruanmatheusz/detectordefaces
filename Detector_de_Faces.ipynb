```python
from mtcnn import MTCNN
from matplotlib import pyplot
from matplotlib.patches import Rectangle
```

```python
img = pyplot.imread('pessoas.jpg')
```

```python
detector = MTCNN()
faces = detector.detect_faces(img)
```

```python
faces
```

```python
len(faces)
```

```python
def caixas(img, faces):
  ax = pyplot.gca()

  for face in faces:
    x, y, width, height = face['box']
    rect = Rectangle((x,y), width, height, color='blue', fill=False, lw=2)
    ax.add_patch(rect)

  pyplot.imshow(img)
  pyplot.savefig('faces.png')
  pyplot.show()
```

```python
caixas(img, faces)
```
